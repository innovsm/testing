```python
# testing code 
import base64
def push_file_to_github(owner, repo,sha,  path, code_content, commit_message, token):
    try:
        # Read and encode the file to Base64
        content = base64.b64encode(code_content.encode('utf-8')).decode('utf-8')
        if sha is None:
            print("Failed to get file SHA")
            return

        # Create payload
        data = {
            "message": commit_message,
            "content": content,
            "sha": sha
        }

        # Push using API
        url = f"https://api.github.com/repos/{owner}/{repo}/contents/{path}"
        headers = {"Authorization": f"token {token}", "Accept": "application/vnd.github.v3+json"}

        response = requests.put(url, json=data, headers=headers)

        if response.status_code == 200 or response.status_code == 201:
            return "Success"
        else:
            return "Failure"
    except Exception as e: 
        return e  # returning the main error
```

```python
def longestPalindrome( s: str) -> int:
        odd_count = [0]
        even_count = []
        for i in set(s):
            count_data = s.count(i)
            if(count_data % 2 == 0):
                even_count.append(count_data)
            else:
                odd_count.append(count_data)
        
        print(odd_count, even_count)
        max_odd = max(odd_count)
        sum_even = sum(even_count)
        return max_odd + sum_even
        
```
